String Indexing

```python
name = "Vashnvi"

nameshort = name[0:4] #start from index 0 all the way till 3(excluding 3)
print(nameshort)

character1 = name[4]
print(character1)
```

Slicing

```python
name = "Vashnavi"
print(name[-4:-1])
print(name[1:4])
print(name[:4])#is same as print[0:4]
print(name[1:])#is same as print[1:5]
print(name[1:5])
```

Skipping

```python
a = "1234567890"
print(a[1:9:2])
print(a[1:5])

w = "amazing"
print(w[:7])
print(w[0:])
```

String Functions

1. Lenght = len() returns the length of the string.

```python
str = "vashnavi"

print(len(str))
```

2. Endswith = endswith() checks if a string ends with given text.

```python
str = "vashnavi"
print(str.endswith("navi"))
print(str.endswith("ere"))
print(str.startswith("vas"))
```

3.  Count = count() counts total occurrences of a character.


```python
name = "vashnavi"
count = name.count("a")
print(count) 
```

4. Capitalize = capitalize() capitalizes the first character.

```python
str = "vashnavi"
capitalized = str.capitalize()
print(capitalized) 
```

5. Find = find() returns the index of first occurrence.

```python
str = "Vashnavi"
index = str.find("h")
print(index) # Output: 2
```

6. Replace = replace(old word, new word) replaces the old word with the new word in the string.


```python
str = "vashnavi"
replaced = str.replace("a", "l")
print(replaced) 
```
