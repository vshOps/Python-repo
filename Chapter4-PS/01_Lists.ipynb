```python
friends = ['Apple', 'Banana', 'Mango', 'Grapes', 1, 0.589]
print(friends[4])

friends[0] = 'Pineapple' #unlike strings lists arte mutable

print(friends[0])
```

```python
friends.append('Watermelon') #add an element at the end of the list
print(friends)
```

```python
l1 = [1, 5, 2, 4, 3, 9, 7, 6, 8]
l1.sort() #sort the list in ascending order
print(l1)
```

```python
l1.reverse() #sort the list in descending order
print(l1)   
```

```python
l1.insert(4, 22222) #insert an element at a specific index
print(l1)
```

```python
print(l1.pop(3))
print(l1) #remove an element at a specific index
```

```python
print(l1.remove(2)) #remove an element by value
print(l1) 
```
